```python
# Lets try on another dataset
import pandas as pd
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Create a histogram of age
heart_disease["age"].plot.hist(bins=50);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
## which one should you use? pyplot bs matplotlibmethod
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
len(over_50)
```

```python
over_50.plot(kind='scatter', x= 'age', y='chol')
```

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target',ax=ax);
ax.set_xlim([45,100])
```

```python
# OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],c=over_50["target"]);
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel='Age', ylabel='Cholesterol');

ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
# subplot of chol,age,thalach
fig, (ax0,ax1) = plt.subplots(nrows=2, ncols=1,figsize=(10,10),sharex=True)
# add data tto ax0 
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],
                     c=over_50["target"])
ax0.set(title= "Heart Disease and Cholesterol levels",
        ylabel="Cholesterol");
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");
scatter = ax1.scatter(x=over_50["age"], y= over_50["thalach"],
                      c=over_50["target"])
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel= "Age",ylabel="Max Heart Rate");
ax1.legend(*scatter.legend_elements(), title ="Target")
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");
 # add a title to fig
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
